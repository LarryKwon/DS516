## merged_event에서 확인할 수 있는 이벤트 타입

```python
df = pd.read_csv("./merged_events.csv")

# Group by 'type' and collect non-empty column names per type
type_columns = {}
for event_type, group in df.groupby('type'):
    non_empty_columns = group.dropna(axis=1, how='all').columns.tolist()
    type_columns[event_type] = non_empty_columns

print(type_columns.keys())
print(type_columns['CHAMPION_KILL'])
```

## participant frames 합치기

```python
import pandas as pd
from pathlib import Path

# Define directory where participant frame CSVs are stored
csv_dir = Path("./participant_frames_output/")

# Ensure the directory exists (if not, likely the user hasn't created it yet)
if not csv_dir.exists():
    merged_df = pd.DataFrame()  # Empty if nothing found
else:
    # Collect all CSV files from the directory
    csv_files = list(csv_dir.glob("*_participant_frames.csv"))

    # Load and concatenate them all
    dfs = [pd.read_csv(f) for f in csv_files]
    merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged result
merged_output_path = csv_dir / "all_participant_frames_merged.csv"
merged_df = merged_df.sort_values(by=['gameId', 'timestamp','participantId'])
merged_df.to_csv(merged_output_path, index=False)
```
